```python
import matplotlib.pyplot as plt
import numpy as np

# 架构与门类型
architectures = ['Superconductor', 'Ion Trap']
gate_types = ['CNOT', 'Hadamard', 'RX', 'PauliRot', 'RZ', 'SWAP']

# 你提供的 gate counts
superconductor_counts = {'CNOT': 48, 'Hadamard': 36, 'RX': 32, 'PauliRot': 30, 'RZ': 8, 'SWAP': 4}
iontrap_counts =        {'CNOT': 48, 'Hadamard': 36, 'RX': 32, 'PauliRot': 30, 'RZ': 8, 'SWAP': 0}

# 添加 total 项
superconductor_total = sum(superconductor_counts.values())
iontrap_total = sum(iontrap_counts.values())

# 插入 "Total" 门数
gate_types_with_total = gate_types + ['Total']
superconductor_values = [superconductor_counts.get(g, 0) for g in gate_types] + [superconductor_total]
iontrap_values =        [iontrap_counts.get(g, 0) for g in gate_types] + [iontrap_total]

# 绘图
x = np.arange(len(gate_types_with_total))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))
bar1 = ax.bar(x - width/2, superconductor_values, width, label='Superconductor')
bar2 = ax.bar(x + width/2, iontrap_values, width, label='Ion Trap')

ax.set_xlabel('Gate Type')
ax.set_ylabel('Gate Count')
ax.set_title('Gate Count Comparison: Superconductor vs Ion Trap (2 FALQON Layers)')
ax.set_xticks(x)
ax.set_xticklabels(gate_types_with_total)
ax.legend()
ax.grid(True, axis='y')

# 每根柱子上标上数值
for bars in [bar1, bar2]:
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{int(height)}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```
