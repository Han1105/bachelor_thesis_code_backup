```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Online Education Market Growth (2023–2025)
years = [2023, 2024, 2025]
market_size = [4.4, None, 6.2]  # Intermediate value to be filled using CAGR
cagr = 0.19

# Fill in 2024 value using CAGR
market_size[1] = round(market_size[0] * (1 + cagr), 2)

plt.figure(figsize=(8, 6))
plt.plot(years, market_size, marker='o')
plt.title('China Online Education Market Growth (2023–2025)')
plt.xlabel('Year')
plt.ylabel('Market Size (Billion USD)')
plt.grid(True)
plt.show()

# 2. Education Expenditure by City Tier (2023)
tiers = ['First-tier Cities', 'Third/Fourth-tier Cities']
expenditure = [1000, 600]

plt.figure(figsize=(8, 6))
plt.bar(tiers, expenditure, color=['#4daf4a', '#377eb8'])
plt.title('Average Annual Education Expenditure by City Tier (2023)')
plt.xlabel('City Tier')
plt.ylabel('Average Expenditure (USD)')
plt.show()

# 3. Growth Rate Comparison of Education Investment by City Tier
growth_rates = [8, 15]
x = np.arange(len(tiers))

plt.figure(figsize=(8, 6))
plt.bar(x, growth_rates, color=['#ff7f00', '#984ea3'])
plt.xticks(x, tiers)
plt.title('Education Investment Growth Rate by City Tier')
plt.xlabel('City Tier')
plt.ylabel('Growth Rate (%)')
plt.show()
```
